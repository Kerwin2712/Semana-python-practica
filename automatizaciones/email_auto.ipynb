## 1.1. Obtener datos de acciones

```python
import yfinance

#ticker = input("Ingrese el ticker de la empresa: ")
ticker = "AAPL"
#ticker = "MSFT"
#ticker = "GOOG"

#dt_inicio = input("Ingrese la fecha inicial en formato YYYY-MM-DD: ")
#de_final = input("Ingrese la fecha final en formato YYYY-MM-DD: ")
dt_inicio = "2022-01-01"
de_final = "2022-01-31"
#dt_inicio = "2022-01-01"
#de_final = "2022-01-31"
#dt_inicio = "2022-01-01"
#de_final = "2022-01-31"
#dt_inicio = "2022-01-01"
#de_final

datos = yfinance.Ticker(ticker)
tabla = datos.history(start=dt_inicio, end=de_final)
#datos = datos.history(start="2022-01-01", end="2022-01-31")
#datos = datos.history(start="2022-01-01", end="2022-01-31")
#datos = datos.history(start="2022-01-01", end="2022-01-31")
#datos = datos.history(start="2022-01-01", end="2022-01-31")
print(tabla)
```

```python
cierre = tabla.Close
cierre
```

```python
cierre.plot(title="Cierre de la acciÃ³n", ylabel="Precio de cierre", xlabel="Fecha")
```

```python
maxima = round(cierre.max(), 2)
minima = round(cierre.min(), 2)
media = round(cierre.mean(), 2)

print(f"El precio mÃ¡ximo fue de {maxima}")
print(f"El precio mÃ­nimo fue de {minima}")
print(f"La media de precios fue de {media}")
```
